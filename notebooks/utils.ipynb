### Helper Functions

```python
"""
Converts a dbfs:/Workspace/... path to its local Linux equivalent.
CE-safe workaround for checkpoint directories.
"""
def workspace_to_local_path(workspace_dbfs_path: str) -> str:
    if workspace_dbfs_path.startswith("dbfs:/Workspace/"):
        # Databricks CE cannot use dbfs:/Workspace/, so map to /Workspace/
        local_path = workspace_dbfs_path.replace("dbfs:", "")
    elif workspace_dbfs_path.startswith("/Workspace/"):
        local_path = workspace_dbfs_path
    else:
        local_path = f"/Workspace/Users/{current_user}/Files/retail_cdc_demo/data/checkpoints"
    return local_path
```
